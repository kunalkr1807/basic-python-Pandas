```python
#  1. access the value within a dataframe . Eg row c , cloumn t , then o/p = ct ,its intersect value
q=matr1.values[matr1.values<0]
for i in matr1.columns:
    for j in matr1.index:
        if matr1[i][j] in q:
            print(i,j,matr1[i][j])
        
```

```python
# 2nd way to achieve the value within a DF.Eg row c , cloumn t,then o/p = ct ,its intersect value
for i in matr1.columns:
    for j in matr1.index: 
        if matr1.loc[j,i]<0: # here we wrote j before i bcoz j is row,i is column.operation starts with row*column
            print(i,j,matr1.loc[j,i])
            
```

```python
df0={'dek1':['d1','d2','d3','d4',],'dek2':[1,0,3,6],'dek3':['qw','op','rg','bh']}
df1=pn.DataFrame(df0)
df1
```

```python
df01={'dek1':['d1','d2','d0','d4',],'dek2':[1,0,3,6],'dek3':['qw','op','rg','bh']}
df2=pn.DataFrame(df01)
df2
```

NOTE : 
1. In Pandas we use 'merge()' Function to join DataFrames ,
Eg : merged_df = pd.merge(df1, df2, on='key', how='inner/outer/left/right')
2. 'on' - specifies the column(s) to use as the key for the join.
3. 'how'- it specifies the type of join to be performed:
3.1. 'inner': returns only the rows that have matching keys in both DataFrames (default).
3.2. 'outer': returns all rows from both DataFrames.
3.3. 'left': returns all rows from the left DataFrame and only the matching rows from the right DataFrame.
3.4. 'right': returns all rows from the right DataFrame and only the matching rows from the left DataFrame.


```python
pn.merge(df1,df2,on='dek1',how='outer')
```

```python
pn.merge(df1,df2,on=['dek1','dek2'])
```

```python
df3 = pn.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

df4 = pn.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

output is same in both merge and join, difference is way of accessing the data.

1. we use column to access the data in 'merge' Function.
2. And we use index in case of joining.

```python
df3.join(df4,how='left')
```
